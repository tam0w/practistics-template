## Practistics Debugging Notebook

```python
import pickle, requests, io, cv2 as cv, matplotlib.pyplot as plt
```

```python
error_id = input("Enter the error id: ")
local_server = True if input("Is the server running locally? (y/n): ") == 'y' else False
```

```python
url = f"http://localhost:5000/api/errors/{error_id}" if local_server else f"https://practistics.live/api/errors/{error_id}"
headers = {'Admin': 'True'}
response = requests.post(url, headers=headers)
print(response)
```

```python
# Get the content of the response
content = response.content

# Create a BytesIO object from the content
byte_io = io.BytesIO(content)

# Unpickle the data
data = pickle.load(byte_io)
data['error_data']['summary']
```

```python
plt.imshow(data['error_data']['summary'])
plt.show()
```
